#Based on filtering and value counts

**How many no. of male and female were driving at a high speed**

```python
df[df.violation == 'Speeding'].driver_gender.value_counts()
```

**No. of males and females who gets searched on a stop**

```python
df.groupby('driver_gender').search_conducted.sum()
```

#or

**In pandas, agg is a method that is short for "aggregation." It is used to perform aggregation operations on a DataFrame, allowing you to apply one or more aggregation functions to specific columns or groups of data. Aggregation functions are used to compute a summary statistic (e.g., sum, mean, count) on a set of values.**

```python
df.groupby('driver_gender')['search_conducted'].agg('sum')
```

**No. of times the searched has been conducted**
